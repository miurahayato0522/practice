```python
import numpy as np
import cv2
import PIL.ImageGrab as ImageGrab
import keyboard
import csv
from paddleocr import PaddleOCR

ocr = PaddleOCR(use_angle_cls=True, lang='en')

left = 236
top = 411
right = 733
under = 527

# CSV ファイルのヘッダーを設定
csv_header = ['OCR_output']

# CSV ファイルを開き、ヘッダーを書き込む
with open('output.csv', 'w', newline='') as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerow(csv_header)

while True:
    box = (left, top, right, under) 
    screen = ImageGrab.grab(box)
    screen = cv2.cvtColor(np.array(screen), cv2.COLOR_RGB2BGR)
    cap = screen
    ret, img = 1, cap

    cv2.imshow('img', img)
   
    # エンターキーが押された瞬間のみ OCR 実行
    if keyboard.is_pressed('enter'):
        result = ocr.ocr(img)
        number_text = result[0][0][1][0] if result and result[0] and result[0][0] and len(result[0][0]) >= 2 else ''
        print(result)
        
        # CSV ファイルに結果を追記
        with open('output.csv', 'a', newline='', encoding='utf-8') as csvfile:
            csv_writer = csv.writer(csvfile)
            csv_writer.writerow([number_text])

    k = cv2.waitKey(30) & 0xff
    if k == 27:
        break

cv2.destroyAllWindows()
```

```python
import pyautogui
import keyboard
import time

def get_mouse_position():
    # マウスカーソル位置の座標を取得する関数
    x, y = pyautogui.position()
    return x, y

def main():
    print("enterキーで座標取得、escキーで終了")
    time.sleep(0.1)
    
    enter_pressed = False

    while True:
        # enterキーが押された瞬間のマウスカーソル位置表示
        if keyboard.is_pressed("enter") and not enter_pressed:
            x, y = get_mouse_position()
            print(f"Mouse Position: ({x}, {y})")
            enter_pressed = True
        else:
            enter_pressed = False

        # escキーで終了
        if keyboard.is_pressed("esc"):
            print("終了")
            break

        # 一度押されたら0.1秒待機用
        time.sleep(0.1)

if __name__ == "__main__":
    main()
```
